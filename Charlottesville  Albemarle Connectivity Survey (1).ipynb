Get dummy for features with more than one choice

```python
def parseMultipleColum(input):
    dummy_7=input.str.join(sep='').str.get_dummies(sep=', ')
    other,nonOther=[],[]

    for col in dummy_7:
        if dummy_7[col].sum()==1:
            other.append(col)
        else:
            nonOther.append(col)

    dummy_7_other=dummy_7[other]
    dummy_7_nonOther=dummy_7[nonOther]
    dummy_7_nonOther['others']= dummy_7_other.sum(axis=1)
    return dummy_7_nonOther,other
```

```python
def demoAna(dummy_1to6,index=-4):
    demo=dummy_1to6.columns[:index]
    freq=dummy_1to6.columns[index:]
    demo_res=[]
    for i in demo:
        res=[]
        for j in freq:
            work_sum=dummy_1to6.loc[dummy_1to6[i] == 1, j].sum()
            res.append(work_sum)
        demo_res.append(res)
    
    demo_df=pd.DataFrame(demo_res)
    demo_df.columns=freq
    demo_df['demographic']=demo
    demo_df=demo_df.set_index('demographic').T
    
    return demo_df.iloc[:, :].apply(lambda x: x / x.sum()).T
    
```

```python
demo_work=demoAna(dummy_1to6)
demo_work
```

```python
import string
import operator

res=[]
stopwords=['AWESOME','Also','Almost','Better','Connect','Connecting','Bikes']

for sentence in file['sugg_conn']:
    words=str(sentence).split(' ')
    for word in words[1:]:
        if len(word)>4 and word not in stopwords:
            if word[0].isupper() or '0'<=word[0]<='9':
                word=word.translate(None,string.punctuation)
                res.append(word)
                
                
               
location={}
for i in res:
    if i in location.keys():
        location[i]+=1
    else:
        location[i]=1

location = {key:val for key, val in location.items() if val != 1}
sorted_location = sorted(location.items(), key=operator.itemgetter(1))
sorted_location[::-1][:20]
   
```

```python
import nltk
stopwords = nltk.corpus.stopwords.words('english')

sugg={}
for sentence in file['sugg_better_conn']:
    words=str(sentence).split(' ')
    for word in words:
        word=word.translate(None,string.punctuation)
        if word in sugg.keys():
            sugg[word]+=1
        else:
            sugg[word]=1
            
for key,val in sugg.items():
    if key.lower() in stopwords:
        del sugg[key]
    
sorted_sugg= sorted(sugg.items(), key=operator.itemgetter(1))
sorted_sugg[::-1][2:30]
```
