```python
#import data and rename columns
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

file=pd.read_csv('C:\Users\yebod\Downloads\survey.csv',skiprows=1,header=None)
file=file.rename(columns={1:'Home',2:'Work',3:'gender',4:'age',5:'race',6:'freq_trans',7:'chanllenge_tran',
                          8:'freq_relax',9:'chanllenge_relax',10:'attri_high_quality',11:'attri_safe',\
                          12:'sugg_conn',13:'sugg_better_conn'})
file.columns
file=file.iloc[5:,1:]
file.head()
```

```python
import nltk

stopwords = nltk.corpus.stopwords.words('english')
```

```python
#build my own stop words to trim useless words
import string
import operator

res=[]
possible_location=['square','hospital','school']
possible_useless=['bike','bikes','trail','connect','connection','need','better','sidewalks',\
                 'sidewalk','safe','would','also','pave','paved','st','street','think','rd',\
                 'trails','road','ave']

word_res=[]
for sentence in file['sugg_conn']:
    if type(sentence)!=str:
        continue
    clean_sent=[]
    
    for word in sentence.split(' '):
        word=word.translate(None,string.punctuation)
        if word!='' and \
        (word[0].isupper() or word.lower() in possible_location or 'th' in word)\
        and word.lower() not in stopwords\
        and word.lower() not in possible_useless:
            clean_sent.append(word)
            word_res.append(word)
    
    res.append(clean_sent)
res[:10]
```

```python
location={}
for i in word_res:
    if i in location.keys():
        location[i]+=1
    else:
        location[i]=1

location = {key:val for key, val in location.items() if val != 1}
sorted_location = sorted(location.items(), key=operator.itemgetter(1))
sorted_location[::-1][:-10]
```

```python
loc_list=location.keys()
loc_out=[x+',charlottsville' for x in loc_list ]
out=pd.DataFrame({'location':loc_out})
out.to_csv('loc_out.csv')
```
